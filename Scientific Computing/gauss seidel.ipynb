```python
import numpy as np
```

```python
a = [
    [6,3,2],
     [2,7,3],
     [7,5,15]
]

print(np.sum(a,axis=1))

def check_diagonally_dominant(a):
  a = np.array(a)

# print(np.diag(a))

# cari diagonal
  diag = np.diag(np.abs(a))

# jumlah tanpa nilai diagonal
  sum_without_diag = np.sum(np.abs(a),axis=1) - diag
  # axis -> untuk nentuin sum arah ke mana
  # axis = 1 -> ke samping
  # axis = 0 -> ke bawah

  if np.all(diag > sum_without_diag):
    print('matrix is diagonally dominant')
  else:
    print('matrix is not diagonally dominant')

```

```python
a = [
    [6,3,2],
     [2,7,3],
     [7,5,15]
]

check_diagonally_dominant(a)
```
