```python
import numpy as np
import matplotlib.pyplot as plt


def original_function(x):
    return np.sin(x)


def nth_derivative_of_original_function(x, n):
    return {
        0: np.sin(x),
        1: np.cos(x),
        2: -np.sin(x),
        3: -np.cos(x)
    }[n % 4]


def taylor_series_approximatio(x, x0, n_terms):
    terms = [nth_derivative_of_original_function(x0, i) * (x - x0)**i / np.math.factorial(i) for i in range(n_terms)]
    return np.sum(terms, axis=0)


def plot_approximation(x_values, approximated_values, n_terms, x0):
    plt.figure(figsize=(10, 6))
    plt.plot(x_values, original_function(x_values), label='Original Function (sin(x))', color='blue')
    plt.plot(x_values, approximated_values, label=f'Taylor Series Approximation (n={n_terms})', color='red', linestyle='--')
    plt.title(f'Taylor Series Approximation of sin(x) around x0={x0}')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)
    plt.show()


x0 = 0.0  
x_values = np.linspace(-2*np.pi, 2*np.pi, 100)  
n_terms = 5  


approximated_values = taylor_series_approximation(x_values, x0, n_terms)

plot_approximation(x_values, approximated_values, n_terms, x0)
```
