```python
# import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# array biasa
a = [1,2,3,4,5]
b = [6,7,8,9,10]

# ubah array biasa ke dalam bentuk array numpy
x = np.array(a)
y = np.array(b)

# plot titik x dan y
plt.plot(x,y,'bx')
plt.show()

```

```python
# data array
# a = [1,2,3,4,5]
# b = [6,7,8,9,10]

# x = np.array(a)
# y = np.array(b)

# persamaan garis lurus
# y = mx + c
m = 1
c = 5

#menampilkan titiknya
plt.plot(x,y,'go')

# menampilkan garis lurusnya
plt.plot(x, m*x+c,'r')
```

```python
# (A^T * A)^-1 A^TY

# data array
a = [1,2,3,4,5]
b = [6,7,8,9,10]

x = np.array(a)
y = np.array(b)

A = np.vstack((x,np.ones(len(x)))).T

y = y[:, np.newaxis]

# di-cut

alpha = np.dot(np.dot(np.linalg.inv(np.dot(A.T, A)), A.T), y)

print(alpha)

# alpha udah ngebungkus
# index 1 = m
# index 2 = c

plt.plot(x,y,'g.')
plt.plot(x, alpha[0]*x + alpha[1])
plt.show()
```
