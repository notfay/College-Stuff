```python
import numpy as np
```

```python
def f(x) :
    return 3*x**2 + 4
```

```python
def g(x) :
    return 6*x
```

```python
def newtonR(x0, tolerance = 0.034, i = 1) :
    if i >= 30 :
        print("EROOOOOOOOOOOOOOOOOR")

    x1 = x0 - f(x0) / g(x0)

    if np.abs(f(x1)) < tolerance :
        print("Root = ", x1)
        return 
    else : 
        print(f"Iteration {i}. new root = {x1}")

    newtonR(x1, tolerance, i+1)
```

```python
newtonR(4)
```
