```python
import Bio
import matplotlib.pyplot as plt
from Bio.Seq import Seq
from Bio import SeqIO
```

# Fasta

```python
# Fasta Reader

for fastaRecord in SeqIO.parse('./Assets/sequence.fasta', 'fasta') :
    print(f" ID: {fastaRecord.id}")
    print(f" Desc: {fastaRecord.description}")
    print(f" Seq: {fastaRecord.seq}")
```

```python
seqA = fastaRecord.seq
print(seqA)
```

```python
# Read 1 Sequence

x = SeqIO.read('./Assets/sequence.fasta', 'fasta')
print(x.seq)
```

# GenBank

```python
for gbRecord in SeqIO.parse('./Assets/sequence.gb', 'gb') :
    print(f" ID: {gbRecord.id}")
    print(f" Desc: {gbRecord.description}")
    print(f" Seq: {gbRecord.seq}")
```

# Splicing

```python
seq1 = seqA[:round(len(seqA) / 2)] # First Half
seq2 = seqA [round(len(seqA) / 2):] # Second Half

print(seq1)
print(seq2)
```

# Melting Content

```python
from Bio.SeqUtils import GC
from Bio.SeqUtils import MeltingTemp as mt

# GC Content

gcA = GC(seq1)
print(gcA)

gcB = GC(seq2)
print(gcB)

# MT

print("---------------")
mtA = mt.Tm_NN(seq1)
mtB = mt.Tm_NN(seq2)

print(mtA)
print(mtB)
```

# Plotting

```python
fig, axs = plt.subplots(1, 2)

# GC Content

axs[0].bar(['GC Seq A', 'GC Seq B'], [gcA, gcB])
axs[0].set_title('GC Content')

# Melt Temp

axs[1].bar(['MT Seq A', 'MT Seq B'], [mtA, mtB])
axs[1].set_title('MT Content')
```
