```python
# Import Module

from Bio.Seq import Seq
from Bio import SeqIO
```

```python
#cara 1 pake parse
for record in SeqIO.parse('./Assets/sequence.fasta','fasta'):
    print(record.id)
    print(record.description)
    print(record.seq)
```

```python
#cara 1 pake parse
for record in SeqIO.parse('./Assets/sequence.gb','gb'):
    print(record.id)
    print(record.description)
    print(record.seq)
```

```python
#cara 2 pakai read
x = SeqIO.read('./Assets/sequence.fasta','fasta')
print(x)

y = SeqIO.read('./Assets/sequence.gb','gb')
print(y.seq)
```

```python
#slice sequence jadi 2
seqX = x.seq
print(seqX)

seqA = seqX[:round(len(seqX)/2)] #bagian awal
seqB = seqX[round(len(seqX)/2):] #bagian akhir

seq10pertama = seqX[:10]
seq10terakhir = seqX[-10:]

print(seq10pertama)
print(seq10terakhir)

print(seqA)
print(seqB)
```

```python
from Bio.SeqUtils import GC

gcA = GC(seqA)
gcB = GC(seqB)
atA = 100 - gcA
atB = 100 - gcB

print(gcA)
print(gcB)
print(atA)
print(atB)
```

```python
from Bio.SeqUtils import MeltingTemp as mt, molecular_weight

mtA = mt.Tm_Wallace(seqA)
mtB = mt.Tm_Wallace(seqB)

mwA = molecular_weight(seqA)
mwB = molecular_weight(seqB)

print(mtA)
print(mtB)
print(mwA)
print(mwB)
```

```python
from matplotlib import pyplot as plt
```

```python
#SubPlot
fig, axs = plt.subplots(1,3)

axs[0].bar(['GC A', 'GC B', 'AT A', 'AT B'], [gcA, gcB, atA, atB])
axs[0].set_title('GC AT Content')

axs[1].bar(['MT A', 'MT B'], [mtA, mtB])
axs[1].set_title('Melting Temp')

axs[2].bar(['MW A', 'MW B'], [mwA, mwB])
axs[2].set_title('Molecular Weight')

plt.tight_layout()
plt.show()
```
