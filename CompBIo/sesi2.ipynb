```python
from Bio.Seq import Seq as seq
import matplotlib.pyplot as plt
```

```python
seq_a = seq('AGCTTGCAGCGTCCGTTAGCTCGAGTCCAGGACGTTAGTCCTGCAGTC')
seq_b = seq('CAGTAAGTTGCCGTTAGCGCGTAGTGCCAGTAAGCGGCTCGTTAGTGG')
```

GC and AT Composition


```python
from Bio.SeqUtils import GC as gc

gc_content = gc(seq_a)
gc_content2 = gc(seq_b)

print(f'Sequence A GC {gc_content:.2f}')
print(f'Sequence B GC {gc_content2:.2f}')
fig, axis = plt.subplots(1, 2, figsize=(12,6))

axis[0].bar(
    ["Seqiemce A GC"],
    [gc_content]
)          
axis[1].bar(
    ["MT Wallace, Sequence B"],
    [gc_content2]
)
```

```python
#AT Content
at_content = 100 - gc_content
at_content2 = 100 - gc_content2

print(f'Sequence A AT {at_content:.2f}')
print(f'Sequence B AT {at_content2:.2f}')

fig, axis = plt.subplots(1, 2, figsize=(12,6))
axis[0].bar(
    ["Seqiemce A AT"],
    [at_content]
)          
axis[1].bar(
    ["Sequence B AT"],
    [at_content2]
)
```

Melting Point of DNA

```python
from Bio.SeqUtils import MeltingTemp as mt

#Melting temp of the DNA Structure from double helix to single

#Calculate MT based of GC, NN, Wallace, thermodynamic
```

```python
#MT based of GC, NN

print(f'Melting point GC, Sequence A : {mt.Tm_GC(seq_a):.2f}')
print(f'Melting point NN, Sequence A : {mt.Tm_NN(seq_a):.2f}')
print(f'')
print(f'Melting point GC, Sequence B : {mt.Tm_GC(seq_b):.2f}')
print(f'Melting point NN, Sequence B : {mt.Tm_NN(seq_b):.2f}')
```

```python
#MT based of Wallace, Thermodynamic

print(f'Melting point Wallace, Sequence A : {mt.Tm_Wallace(seq_a):.2f}')
print('')
print(f'Melting point Wallace, Sequence B : {mt.Tm_Wallace(seq_b):.2f}')

mt_W = mt.Tm_Wallace(seq_a)
mt_W2 = mt.Tm_Wallace(seq_b)

fig, axis = plt.subplots(1, 3, figsize=(12,6))

axis[0].bar(
    ["MT Wallace, Sequence A"],
    [mt_W]
)          
axis[1].bar(
    ["MT Wallace, Sequence B"],
    [mt_W2]
)
axis[2].pie(
    [mt_W, mt_W2],
    labels=[f'Sequence A {mt_W:.0f}%', f'Sequence B {mt_W2:.0f}%']
)
axis[2].legend(["MT Wallace Sequence A", "MT Wallace Sequence B"], loc="upper right")
```

```python
#MT based of Thermodynamic
thermodynamic = mt.Tm_NN(
    seq_a,
    Na = 50,
    Mg = 10
)

thermodynamic2 = mt.Tm_NN(
    seq_b,
    Na = 50,
    Mg = 10
)

print(f'Thermodynamic Sequence A = {thermodynamic:.2f}')
print(f'Thermodynamic Sequence B = {thermodynamic2:.2f}')
```

Molecular Weight

```python
from Bio.SeqUtils import molecular_weight as mw

mwA = mw(seq_a)
mwB = mw(seq_b)


print(f'Sequence A MW {mwA:.2f}')
print(f'Sequence B MW {mwB:.2f}')
```
